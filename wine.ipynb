```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.losses import SparseCategoricalCrossentropy
from keras.layers import Dense
```

```python
data=pd.read_csv("WineQT.csv")
data.head()
```

```python
data=data[['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol','quality']]
data.head()
```

```python
model=Sequential([
    Dense(units="256",activation="relu",name="Layer_1",kernel_regularizer=tf.keras.regularizers.L2(0.001)),
    Dense(units="256",activation="relu",name="Layer_2",kernel_regularizer=tf.keras.regularizers.L2(0.001)),
    Dense(units="128",activation="relu",name="Layer_3",kernel_regularizer=tf.keras.regularizers.L2(0.001)),
    Dense(units="11",activation="softmax",name="Layer_4",kernel_regularizer=tf.keras.regularizers.L2(0.001)),
])
model.compile(loss=SparseCategoricalCrossentropy(),optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),metrics=['accuracy'])
```

```python
from sklearn.model_selection import train_test_split
X=data[['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol']].values
Y=data['quality'].values.reshape(-1)
x_train,x_test,y_train,y_test=train_test_split(X,Y,train_size=0.8,random_state=42)
x_train,x_cv,y_train,y_cv=train_test_split(x_train,y_train,random_state=42,train_size=0.8)
```

```python
print(X.shape)
print(Y.shape)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
model.fit(x_train,y_train,epochs=250,batch_size=20,validation_data=(x_cv, y_cv))
```

```python
model.summary()
```

```python
y_pred=model.evaluate(x_test,y_test)
```

```python
model.save('wine.keras')
```
